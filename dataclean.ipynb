```python
##import packages etc.
import pandas as pd
import numpy as np 
from skimpy import clean_columns
```

```python
#import data
df= pd.read_csv('data/Survey.csv')
```

```python
##quick view of data
df.head
```

```python
##list of columns
list(df.columns)
```

```python
#quick attempt at cleaning columns using skumpy/pascal case
cdf=df = clean_columns(df, case= 'pascal')
cdf.columns
```

```python
##cleaning and renaming columns even further maually using 'rename' pandas function
###only renamed the columns that would be used for analytical purposes
cdf.rename(columns={'WhatIsYourName': 'name', 'DoYouWorkInHealthcare': 'workinhealthcare',
                        'AreYouAHealthcareProviderOrAnItProfessional': 'jobcategory', 
                        'WhereDoYouWork': 'worklocation',
                        'IsThisAPrivatePractice': 'privatepractice',
                        'WhatIsYourRoleOrJobDescription': 'role',
                        'ToWhatExtentDoYouBelieveYouUnderstandWhatInteroperabilityIsInHealthItAndHowItPertainsToYourAbilityToCompleteYourJob': 'interoperability understanding',
                        'HowFrequentlyDoYouEncounterChallengesOrDifficultiesInExchangingPatientInformationBetweenDifferentHealthItSystems': 'difficulty frequency',
                        'HowWouldYouRateTheLevelOfInteroperabilityBetweenTheHealthItSystemsUsedInYourOrganization': 'Interoperability Rating',
                        'WhichOfTheFollowingChallengesHaveYouEncounteredWhenExchangingPatientDataBetweenDifferentHealthItSystems': 'challenges',
                        'WhichOfTheFollowingDataStandardsOrProtocolsDoesYourOrganizationPrimarilyUtilizeForInteroperabilityPurposes': 'Health Standards',
                        'WhatEhrSystemDoesYourOrganizationUseIfYouDontKnowTypeIdk': 'EHR',
                        'WhatCloudComputingSystemDoesYourOrganizationUse': 'Cloud Platform',
                        'HowEffectiveAreTheExistingInteroperabilitySolutionsInEnsuringTheSeamlessFlowOfPatientDataAcrossDifferentHealthItSystems': 'Interoperability Effectivenesss',
                        'ToWhatExtentDoYouBelieveThatImprovedInteroperabilityCanEnhanceYourAbilityToProvideSufficientPatientCareCoordinationAndHealthcareOutcomes': 'Need for improvement'},inplace = True)
```

```python
##quick list of new and improved column names for assurance
list(cdf.columns)
```

```python
#removing commas,slashes, quotes and etc. from values within the columns using str.replace
##this is to ease the process of cleaning when converting values to integers

cdf['Health Standards'] = cdf['Health Standards'].str.replace(r"[\"\',]", '')
cdf['Interoperability Effectivenesss'] = cdf['Interoperability Effectivenesss'].str.replace(r"[\"\',]", '')
cdf['Need for improvement'] = cdf['Need for improvement'].str.replace(r"[\"\',]", '')
cdf['interoperability understanding'] = cdf['interoperability understanding'].str.replace(r"[\"\',]", '')
```

```python
#Renaming values to an integer form using remap function
##This will help when analyzing the data
###This will act as our data dictionary 

privatepractice_map = {'Yes': 1, 'No': 0}
cdf.replace({'privatepractice': privatepractice_map}, inplace=True)

jobcategory_map = {'Provider': 1, 'IT Professional': 0, 'It Professional':0}
cdf.replace({'jobcategory': jobcategory_map}, inplace=True) 

interoperabilityunderstanding_map = {'None': 1, 'MInimally': 2, 'Somewhat': 3, 'Moderately': 4, 'Significantly': 5, 'Minimally':2, 'I dont know and/or dont care':0}
cdf.replace ({'interoperability understanding': interoperabilityunderstanding_map}, inplace=True)

InteroperabilityRating_map = {'Horrible': 1, 'Not that good': 2, 'Neutral': 3, 'Good': 4, 'Excellent': 5, }
cdf.replace({'Interoperability Rating': InteroperabilityRating_map}, inplace=True)

Needforimprovement_map = {'None': 1, 'Minimally': 2, 'Somewhat': 3, 'Moderately': 4, 'Significantly': 5, 'significantly': 5, 'NaN': 0, 'IDK':0, 'I dont know and/or dont care':0 }
cdf.replace({'Need for improvement': Needforimprovement_map}, inplace=True)

difficultyfrequency_map = {'Never': 1, 'Not often': 2, 'Sometimes': 3, 'Often': 4, 'Very Often': 5, 'Not Often':2 }
cdf.replace({'difficulty frequency':difficultyfrequency_map}, inplace=True)

HealthStandards_map= {'HL7': 1, 'I dont know': 0, 'DICOM': 2, 'NaN': 0, 'HIPAA': 3}
cdf.replace({'Health Standards': HealthStandards_map}, inplace=True)

CloudPlatform_map= {'Microsoft Azure': 1, 'Microsoft': 1, 'Micosoft Azure':1, 'IDK': 0, 'Google Cloud Computing': 2, 'NaN': 0, 'AWS': 3, 'Not sure':0, 'Not sure, I think outlook':0, 'Microsoft OneDrive':1, 'Epic':0}
cdf.replace({'Cloud Platform': CloudPlatform_map}, inplace=True)

EHR_map= {'EPIC': 1, 'Epic':1, 'I dont know': 0, 'Allscript': 2, 'HealthBridge':3, 'Health Bridge': 3, 'Health Bridge':3, 'MyChart':4, 'IDK':0, 'idK': 0, 'Idk:,0' 'Point Care Manager':5 }
cdf.replace({'EHR': EHR_map}, inplace=True)

InteroperabilityEffectiveness_map = {'Not very effective': 1, 'Somewhat effective': 2, 'Moderately': 3, 'moderately':3, 'Highly': 4, 'I dont know and/or dont care':0, 'IDK':0 }
cdf.replace({'Interoperability Effectivenesss': InteroperabilityEffectiveness_map}, inplace=True)
```

```python
#replacing Nan values with 0
##Nan values will act as an 'IDK' 
cdf= cdf.fillna(0)
```

```python
#Dropping all values that can't assume an integer value or is irrelevant to our study as it pertains to graphing
cdf=cdf.drop(['role', 'name','Timestamp', 'HowWouldYouRateTheLevelOfInteroperabilityBetweenTheHealthItSystemsUsedInYourOrganization1', 
'CanYouDescribeASpecificInstanceWhereYouEncounteredChallengesOrIssuesRelatedToInteroperabilityInYourWork',
'WhatAreTheKeyBarriersOrObstaclesDoYourOrganizationsItSystemsPoseThatHinderTheAchievementOfSeamlessInteroperabilityBetweenHealthItSystemsAndProvidingProperHealthcare',
'InYourOrganizationWhatStepsOrInitiativesHaveBeenTakenToAddressInteroperabilityChallengesPleaseProvideExamplesIfApplicable',
'WhatStepsDoesYourOrganizationTakeInOrderToImproveInteroperability',
'WhatStepsCouldYourOrganizationTakeInOrderToImproveInteroperability'], axis=1)
```

```python
#Saving our new and clean survey csv
cdf.to_csv('data/cleansurvey.csv')
```
